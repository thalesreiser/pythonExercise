```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
dataset=pd.read_csv('/content/gerint_solicitacoes_mod.csv', ';')
dataset
```

```python
def consultMediaIdade(municipio):
  dataMun=dataset[dataset['municipio_residencia']==municipio]
  
  dataMasc=dataMun[dataMun['sexo']=='MASCULINO']
  dataFem=dataMun[dataMun['sexo']=='FEMININO']
  mediaMasc=np.average(dataMasc['idade'])
  mediaFem=np.average(dataFem['idade'])
  mediaIdade=(mediaMasc+mediaFem)/2
  totalPac=dataMun.index.size

  retorno= ['Total de pacientes do município',municipio,
            totalPac,
            ';Média de idade masculino:', mediaMasc,
            ';Média de idade feminina:', mediaFem,
            ';Média de idade de todos pacientes:', mediaIdade]
  return retorno

def consultInterAno(municipio):
  dataMun=dataset[dataset['municipio_residencia']==municipio]

  data2018=dataMun.query('data_internacao=="%2018%"')
  data2019=dataMun.query('data_internacao=="%2019%"')
  data2020=dataMun.query('data_internacao=="%2020%"')
  data2021=dataMun.query('data_internacao=="%2021%"')

  intern2018=data2018.index.size
  intern2019=data2019.index.size
  intern2020=data2020.index.size
  intern2021=data2021.index.size

  internacoes=[intern2018, intern2019, intern2020, intern2021]
  return internacoes

def consultHosp(executante):
  dataExec=dataset[dataset['executante']==executante]

  colunas =['data_extracao',
            'situacao',
            'central_regulacao_origem', 
            'data_solicitacao', 
            'sexo', 
            'solicitante', 
            'codigo_cid', 
            'carater', 
            'tipo_internacao', 
            'tipo_leito', 
            'horas_na_fila']

  dataHosp=dataExec.drop(columns=colunas)

  return dataHosp

def consultCalcIntern(solicitante):
  dataSolic=dataset[dataset['solicitante']==solicitante]

  pacien=dataSolic['id_usuario']

  ex=dataSolic['executante']
  execut=ex.drop_duplicates()
  testeA=pd.Timestamp(dataSolic['data_alta'].values).to_pydatetime()
  testeS=pd.Timestamp(dataSolic['data_solicitacao'].values).to_pydatetime()
  alta=dt.timedelta(testeA)
  solicita=dt.timedelta(testeS)
  quantiDias = abs((alta - solicita).days)
  dataSolic['dias_internacao']=quantiDias
  dataInter=dataSolic['dias_internacao']

  pacien['dias_internacao']=dataInter
  

  return pacien and execut

def consultTempoFila():
  aux=dataset.dropna()
  tempo=aux.sort_values(by='horas_na_fila', ascending=False).head(5)
  ret=np.array(tempo['horas_na_fila'])
  return ret
```

```python
def main():

  while True:
    print('##############SISTEMA DE CONSULTAS################')
    print('Menu de opções: ')
    print('1. Consultar média de idade dos pacientes.')
    print('2. Consultar internações por ano.')
    print('3. Consultar hospitais.')
    print('4. Calcular tempo de internação.')
    print('5. Determinar tempos de espera na fila.')
    print('6. Sair do programa')

    opcao=float (input('Digite a opção do menu: '))

    if opcao==1:
      print('Menu: ', opcao)
      municipio=input('Digite o município que deseja realizar a consulta:')
      print(consultMediaIdade(municipio))
    if opcao==2:
      print('Menu: ', opcao)
      municipio=input('Digite o município que deseja realizar a consulta:')        
      print(consultInterAno(municipio))
    if opcao==3:
      print('Menu: ', opcao)
      executante=input('Digite o executante que deseja realizar a consulta:')
      print(consultHosp(executante))
    if opcao==4:
      print('Menu: ', opcao)
      solicitante=input('Digite o nome do solicitante que deseja realizar a consulta:')
      print(consultCalcIntern(solicitante))
    if opcao==5:
      print('Menu: ', opcao)
      print(consultTempoFila())
    if opcao== 6:
      print('Saindo do sistema!')
      break           
    else:
      print('Opção Inválida! ')

  return 0

main()
```
